```python
import os
import skimage
from skimage import data
import numpy as np
```

```python
def load_data(data_directory):
    
    #Creamos una lista con los directorios que se encuentren en la ruta 
    directories = [d for d in os.listdir(data_directory) if os.path.isdir(os.path.join(data_directory, d))]
    
    #Generamos dos listas vacias para labels e imagenes
    labels = []
    images = []
    
    for d in directories:
        
        #Cargamos una ruta para cada directorio.
        label_directory = os.path.join(data_directory, d)
        #Cargamos un nombre de archivo para cada archivo que sea .ppm 
        file_names = [os.path.join(label_directory, f) for f in os.listdir(label_directory) if f.endswith(".ppm")]
        
        #Por cada archivo cargamos la imagen y el label en las respectivas listas
        for f in file_names:
            
            #Load an image from file
            images.append(skimage.data.imread(f))
            labels.append(int(d))
    
    return images, labels

train_data_directory = "./Training"
test_data_directory = "./Testing"

images, labels = load_data(train_data_directory)
```

```python
#Trabajamos pyplot 
import matplotlib.pyplot as plt

#Creamos un histograma con 62 contenedores.
plt.hist(labels, 62)

plt.show()
```

```python
#Cojemos cuatro indices aleatorios de imagenes que queramos ver
traffic_signs = [300, 2250, 3650, 4000]

for i in range(len(traffic_signs)):
    
    #Pinta una cuadricula con una fila, cuatro columnas y especifica la posicion
    plt.subplot(1, 4, i+1)
    #quitamos los ejes para que solo salgan las fotos y no en forma de gráfico
    plt.axis('off')
    #mostramos las imagenes
    plt.imshow(images[traffic_signs[i]])
    plt.subplots_adjust(wspace=0.5)
    print "Shape: {0}, min: {1}, max {2}.".format(images[traffic_signs[i]].shape, images[traffic_signs[i]].min(),images[traffic_signs[i]].max() )
    
plt.show()
```

```python
#Cojemos los labels
unique_labels = set(labels)

#Inicializamos una nueva figura
plt.figure(figsize=(15,15))

#Establecemos un contador
i = 1

for label in unique_labels:
    #Asignamos la imagen correspondiente 
    image = images[labels.index(label)]
    #Establecemos un subplot de 8 x 8.
    plt.subplot(8,8,i)
    plt.axis('off')
    plt.title("Label {0} ({1})".format(label, labels.count(label)))
    i += 1
    #Mostramos la imagen 
    plt.imshow(image)
    
plt.show()
```

```python
#Podemos observar que hay 375 fotos con el label 22, 316 con el label 32, 285 con el label 38 y 282 con el label 61
#por lo que el histograma que creamos antes es preciso.
```

```python
#Las imagenes no tienen el mismo tamaño, por lo que vamos a hacer un reescalado de estas a 28px mediante skimage
from skimage import transform

images28 = []
for image in images:
    images28.append(transform.resize(image, (28, 28)))
       
```

```python
#Volvemos a convertir la variable que contiene las imagenes a un nparray
images28 = np.array(images28)

#Ahora convertiremos las imagenes a una escala de grises (rgb2gray)
from skimage.color import rgb2gray
images28 = rgb2gray(images28)

#Checkeamos que se haya realizado correctamente seleccionando 4 indices aleatorios para imprimir
traffic_signs = [300, 2250, 3650, 4000]
for i in range(len(traffic_signs)):
    plt.subplot(1, 4, i+1)
    plt.axis('off')
    
    #Tenemos que especificar el comando cmap (mapa de color) ya que por defecto muestra en color.
    plt.imshow(images28[traffic_signs[i]], cmap="gray")
    plt.subplots_adjust(wspace=0.5)
    
# Show the plot
plt.show()
```
